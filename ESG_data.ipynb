```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\emmae\OneDrive\Documents\Data practice\260204_ESGoriginal_v1.csv")
df
```

```python
df.describe()
```

```python
#remove columns

df = df.drop(columns = "MarketCap")
df
```

```python
df = df.drop(columns = "CompanyName")
df
```

```python
#Count of companies from different regions
#would like to show individual companies (no / 11), if not will show as percentages in pie chart

Region_count = df.groupby('Region').agg({'Region':'count'})
Region_count
```

```python
#Count of companies in different industries
#Same as above, would like to filter for exact year

industry_count = df.groupby('Industry').agg({'Industry':'count'})
industry_count
```

```python
#industry chart

Industry_count_chart = pd.DataFrame([1276, 1188, 1243, 1331, 1287,  1166, 1045, 1287, 1177], index=['Consumer Goods', 'Energy', 'Finance', 'Healthcare', 'Manufacturing', 'Retail', 'Technology', 'Transportation', 'Utilities'], columns = ['Industries'])

Industry_count_chart.plot(kind='pie', subplots=True, figsize=(8,8))
```

```python
#region chart

Region_count_chart = pd.DataFrame([1463, 1672, 1540, 1507,1617,1540,1661], index=['Africa','Asia','Europe','Latin America','Middle East','North America','Oceania'], columns = ['Regions'])

Region_count_chart.plot(kind='pie', subplots=True, figsize=(8,8))

```

```python
#average energy consumption over time per industry

Energyconsumption = df.groupby(['Year']).agg({'EnergyConsumption':'mean'})
Energyconsumption.plot(color= 'orange')
```

```python
Energyconsumption = df.groupby(['Industry']).agg({'EnergyConsumption':'mean'})
Energyconsumption.plot(kind= 'bar', color= 'orange')
```

```python
#avg water usage

waterusage= df.groupby(['Year']).agg({'WaterUsage': 'mean'})
waterusage.plot()
```

```python
Waterusage = df.groupby(['Industry']).agg({'WaterUsage':'mean'})
Waterusage.plot(kind= 'bar')
```

```python
#avg carbon emissions

carbonemissions = df.groupby(['Year']).agg({'CarbonEmissions': 'mean'})
carbonemissions.plot(color = 'black')
```

```python
Energyconsumption = df.groupby(['Industry']).agg({'CarbonEmissions':'mean'})
Energyconsumption.plot(kind= 'bar', color= 'black')
```

```python
#revenue over time

Revenueperindustry = df.groupby(['Industry']).agg({'Revenue':'mean'})
Revenueperindustry.plot(kind = 'bar', figsize = (8,8))
```

```python
ESGscore = df.groupby('Year').agg({'ESG_Overall':['mean'],'ESG_Environmental':['mean'],'ESG_Social':['mean'],'ESG_Governance':['mean']})
ESGscore.plot(figsize = (8,8))
```

```python
ESGscoreindustry = df.groupby('Industry').agg({'ESG_Overall':['mean'],'ESG_Environmental':['mean'],'ESG_Social':['mean'],'ESG_Governance':['mean']})
ESGscoreindustry.plot(kind = 'bar',figsize = (8,8))
```

```python
#revenue data

RevenueIndustry = df.groupby(["Industry"]).agg({'Revenue':'mean'})
RevenueIndustry
RevenueIndustry.plot(kind = 'bar')
```

```python
Revenueovertime = df.groupby(["Year"]).agg({'Revenue':'mean'})
Revenueovertime
Revenueovertime.plot()
```

```python
#profit margin

ProfitMarginTime = df.groupby(["Year"]).agg({'ProfitMargin':'mean'})
ProfitMarginTime.plot()
```

```python
ProfitMarginTime = df.groupby(["Industry"]).agg({'ProfitMargin':'mean'})
ProfitMarginTime.plot(kind = 'bar')
```
